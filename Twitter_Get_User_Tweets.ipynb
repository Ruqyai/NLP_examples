```
import tweepy 
import pandas as pd 
import json
from getpass import getpass
import os
```

```
os.environ['consumer_key'] = getpass('Enter the consumer key: ')
os.environ['consumer_secret'] = getpass('Enter the consumer secret: ')
os.environ['access_token'] = getpass('Enter the access token: ')
os.environ['access_secret'] = getpass('Enter the access secret: ')
```

```
auth = tweepy.OAuthHandler(os.environ['consumer_key'], os.environ['consumer_secret'])
auth.set_access_token(os.environ['access_token'], os.environ['access_secret'])
api = tweepy.API(auth, wait_on_rate_limit=True)
```

# User Name

```
# initialize a list to hold all the tweepy Tweets
alltweets = []
User='SaudiMOH'

# make initial request for most recent tweets (200 is the maximum allowed count)
new_tweets = api.user_timeline(screen_name= User, count=200,
                                   tweet_mode='extended')
# save most recent tweets
alltweets.extend(new_tweets)

# save the id of the oldest tweet less one
oldest = alltweets[-1].id - 1

# keep grabbing tweets until there are no tweets left to grab
while len(new_tweets) > 0:
        print("getting tweets before {}".format(oldest))

        # all subsiquent requests use the max_id param to prevent duplicates
        new_tweets = api.user_timeline(screen_name='SaudiMOH', count=200,
                                       max_id=oldest, tweet_mode='extended')

        # save most recent tweets
        alltweets.extend(new_tweets)

        # update the id of the oldest tweet less one
        oldest = alltweets[-1].id - 1

print("...{} tweets downloaded so far".format((len(alltweets))))
```

```
# write tweets in json file
file_json = 'User_Tweets.json'
with open(file_json , 'w') as outfile:
        for t in alltweets:
            outfile.write(json.dumps(t._json))
            outfile.write('\n')
            
            
print('{} tweets have been written successfully'.format((len(alltweets))))
```

```
# show the file and read it
data= []
with open(file_json, 'r') as fh:
    for line in fh:
        tweet = json.loads(line)
        data.append(json.loads(line))
        
df= pd.DataFrame(data)
df
```

```
# save the file as csv file
df.to_csv("User_Tweets.csv")
```
